<font color='brown'><font size="5.9">Part 1. The Introduction of the Redshift-Space Galaxy Power Spectrum</font></font>  

The large-scale structure of the Universe and its evolution over time contains an abundance of cosmological information. 
One way to unlock this is by measuring the density, momentum and cross power spectrum from the positions and peculiar velocities of galaxies, and fitting  the cosmological parameters from these power spectrum. The cosmological parameters that we seek to measure are the growth rate of
structure $$f\equiv  \frac{d\ln\,D(a)}{d\ln\,a}$$ which is the derivative of the linear growth factor $D(a)$ with respect to the  scale factor $a$, and the galaxy bias parameter $$b\equiv  \sqrt{\frac{ P_{g}({\bf k})}{P_{\mathrm{DM}}({\bf k})}}$$ which is the ratio between the galaxies density and dark matter density, where $P_{\mathrm{DM}}$ and $P_{g}$ are the dark matter and galaxies density power spectrum, respectively. Moreover,  the growth rate and galaxy bias parameter can be  normalized as $f\sigma_8$ and $b\sigma_8$, where $\sigma_8$ is defined as  the root mean square of the mass density fluctuation within spheres of 8 Mpc $h^{-1}$. Both $f\sigma_8$ and $b\sigma_8$ are measurable from galaxies. 
    
    
The Hubble recessional redshift and peculiar velocity of a galaxy contribute to form its observed redshift. Therefore, the position of a galaxy inferred from its observed redshift is not its true comoving position (real-space possition). Instead, it is a ‚Äòredshift-space‚Äô position contaminated by the peculiar motion. This is the so-called redshift-space distortion (RSD). The spherical symmetry about the line of sight of the power spectrum estimated from the galaxies‚Äô redshift-space positions is broken by RSD. Therefore, a common method is to decompose the redshift-space power spectrum $P(k,\mu)$ using the Legendre polynomials $ùêø_{\ell}(\mu)$, i.e.
$$P(k, \mu)= \sum_{\ell}P_{\ell}(k )L_{\ell}(\mu)$$
where $\ell=0,1,2,3,4,...$. The $k$ is the wave number. The cosine of the angle between the     unit wave vector ${\bf\hat{k}}$ and the line-of-sight unit position vector ${\bf\hat{r}}$ is $\mu={\bf\hat{r}}\cdot {\bf\hat{k}}=\mathrm{cos}\,\theta$. 
The $P_{\ell}(k )$ is the so-called power spectrum multipoles, which is computed from the Legendre transformation of the
redshift-space power pectrum $P(k,\mu)$, i.e.
$$P_l(k)=(2\ell+1) \int^1_0P(k,\mu)L_l(\mu)d\mu$$
The power spectrum multipoles can be measured from galaxy surveys. By comparing the measured power spectrum multipoles to the theoritical models of power spectrum multipoles, we can estimate the growth rate and bias parameter of the large-scale-structure.  

    
In this notebook, we will explore the galaxy redshift-space auto-density power spectrum $P^{\delta}(k)$, auto-momentum power spectrum $P^{p}(k)$ and the density-momentum-cross power spectrum $P^{\delta p}(k)$. In the absence of wide-angle or general relativistic effects, only the even-multipoles of the auto-power spectrum are non-zero. Under similar conditions, only the odd-multipoles of the cross power spectrum are non-zero.

<font color='brown'><font size="4.9">Section 2.1 Measuring Power Spectrum of Galaxy Survey</font></font>



In this section, we will show how to measure the density, momentum and cross power spectrum of a volume limited galaxy survey.

<font color='brown'><font size="3.9">2.1.1 The Density Power Spectrum</font></font>

 

```python
# the fiducial cosmology
OmegaM = 0.3153     # matter density parameter
OmegaA = 1.-OmegaM  # dark engergy matter density parameter
Hub    = 100.       # Hubble constant
```

<font color='blue'>Step 1. calculate the number ratio between the galaxies and randoms:</font>

```python
D2R_ratio = len(ra)/len(raR)
```

<font color='blue'>Step 2. calculate the number density of galaxies:</font> \
\
we have two different methods to calculate the number density of galaxies $\bar{n}(\vec{r})$.
- calculate $\bar{n}(\vec{r})$ on grid, $\bar{n}(\vec{r})$ is a function of sky coverage (ra, dec) and redshift zm.  
- calculate $\bar{n}(\vec{r})$ in shell, $\bar{n}(\vec{r})$ is a function of redshift zm, it is independent of sky coverage (ra, dec). 

If the sky coverage and completness of a given survey is complex, we strongly recommand to calculate $\bar{n}(\vec{r})$ on grid. If calculating $\bar{n}(\vec{r})$ on grid, the number of randoms should be large enough to avoid $\bar{n}(\vec{r})=0$. In this notebook, we will compare the above two methods.

The galaxy number density $\bar{n}(\vec{r})$ should be computed from the random catalogue, since $\bar{n}(\vec{r})$ should not have any large-scale-structure infomation and poisson noise. See https://arxiv.org/abs/1910.02914 .
To do this, we firstly calculate number density of the random catalogue $\bar{n}_{R}(\vec{r})$: 

<font color='gray'>(if calculating $\bar{n}(\vec{r})$ on grid)</font>\
<font color='green'><font size="2">
nbR_G,NgridDenR,Xlim,Ylim,Zlim = nbarG_Fun(raR,decR,zmR,weight,nbx,nby,nbz,OmegaM,OmegaA,Hub)  
</font></font>

<font color='gray'>(if calculating $\bar{n}(\vec{r})$ in shell)</font> 
<font color='green'><font size="2">    
nbR_S      = nbarS_Fun(delt_z,zmR,weight,SurveyArea,OmegaM,OmegaA,Hub)  
</font></font>

where <font color='green'>weight</font> is the weights of randoms, <font color='green'>delt_z</font> is the thickness of shell, <font color='green'>SurveyArea</font> is the area of the sky coverage of the survey. 
Then calculating galaxy number density $\bar{n}_{g}(\vec{r})$ using $\bar{n}_{R}(\vec{r})$: 

<font color='gray'>(if calculating $\bar{n}(\vec{r})$ on grid)</font>\
<font color='green'><font size="2">
nb_G = nbarGAsign_Fun( ra,dec,zm,  NgridDenR* D2R_ratio,Xlim,Ylim,Zlim, OmegaM,OmegaA,Hub) 
</font></font>

<font color='gray'>(if calculating $\bar{n}(\vec{r})$ in shell)</font> 
<font color='green'><font size="2">    
nb_S       = nbarSAsign_Fun( zm,zmR,     nbR_S * D2R_ratio   ) 
</font></font>

For simplify, we set the weights to be ones.

```python
weit = 1. # setting the weight of galaxies to be 1 for simplify. 


# method-1. calculate nbar on grid: 
nbx,nby,nbz = 128,128,128 # grid size. 
nbR_G,NgridDenR,Xlim,Ylim,Zlim = nbarG_Fun(raR,decR,zmR,weit,nbx,nby,nbz,OmegaM,OmegaA,Hub)# nbar of randoms.
nb_G = nbarGAsign_Fun( ra,dec,zm,  NgridDenR* D2R_ratio,Xlim,Ylim,Zlim, OmegaM,OmegaA,Hub)# nbar of galaxies.
    
    
# method-2. calculate nbar in shell:
delt_z     = 0.002  # the thickness of shell. 
SurveyArea = 4.*np.pi*0.413159  # the area of sky coverage of the survey. 
nbR_S      = nbarS_Fun(delt_z,zmR,weit,SurveyArea,OmegaM,OmegaA,Hub)# nbar of randoms. 
nb_S       = nbarSAsign_Fun( zm,zmR,     nbR_S * D2R_ratio   )# nbar of galaxies. 

print('Warning: if calculate nbar on grid, there are', len(np.where(nb_G==0)[0]),'galaxies with nbar=0. You may need a larger random catalogue!!!')
# make some plots to see how they looks like:
plt.figure(1,figsize=(8,3))
fsiz=15;tsiz=10
plt.subplot(121)
plt.scatter(zm,nb_G ,s=0.1,label='galaxies, using grid');
plt.scatter(zm,nb_S ,s=0.1,label='galaxies, using shell');
plt.xlabel('redshift',fontsize=fsiz);plt.ylabel(r'$\bar{n}$',fontsize=fsiz);plt.tick_params( labelsize =tsiz); plt.legend()
plt.subplot(122)
plt.scatter(zmR[:200000],nbR_G[:200000],s=0.1,label='randoms, using grid');
plt.scatter(zmR[:200000],nbR_S[:200000],s=0.1,label='randoms, using shell');
plt.xlabel('redshift',fontsize=fsiz);plt.ylabel(r'$\bar{n}$',fontsize=fsiz);plt.tick_params( labelsize =tsiz); plt.legend()
plt.subplots_adjust(left=0.125,bottom=0.1, right=1.0, top=0.9, wspace=0.2, hspace=0.23)
plt.show()
```

<font color='red'><font size="3.9">Important Tips:</font></font>  The users are encouraged to calculate the number density of galaxies $\bar{n}(\vec{r})$ using their own methods rather than using our naive algorithm.



<font color='blue'>Step 3. setting k-bins and grid and power spectrum parameters:</font> 

```python
# power spectrum will be measured in k in [0,0.35] for 66 bins. 
kmin,kmax,nk    = 0.0,0.35, 66

# the grid used to grid the galaxies.
nx,ny,nz        = 400,400,400  
xmin,xmax       = -400. ,400. 
ymin,ymax       = -400. ,400.
zmin,zmax       = -400. ,400. 

# FKP weights for density power spectrum. Setting it to be '0' if you do not want any FKP weight.
FKP             = 5000.    

# power spectrum multipoles: 
PS_multi        ='all'  
# 'all'  : will calculate power spectrum for l=0,1,2,3,4
# 'no'   : will calculate power spectrum for l=0 
# 'yes'  : will only calculate the non-zero power spectrum multipoles to save time.
#          i.e. l=0,2,4 for auto-density and auto-momentum power spectrum, 
#               l=1,3   for cross power spectrum.
```

<font color='blue'>Step 4. calculate the density power spectrum:</font> 

```python
# 1. if nbar is calculated on grid: 
# setting PS_type='den' for density power. 
out_dir,wfkpG, NormG = Pkest_Fun(kmin,kmax,nk,nx,ny,nz,xmin,xmax,ymin,ymax,zmin,zmax,
                                ra ,dec ,zm ,nb_G           , FKP,
                                raR,decR,zmR,nbR_G*D2R_ratio, # normalize the nbar of randoms using D2R_ratio. 
                                np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,
                                OmegaM,OmegaA,Hub,np.nan, 
                                np.nan,PS_type='den',PS_multi=PS_multi,
                                file_dir=file_dir+'PSden_grid', wd=1., wdR=1.) # wd and wdR are any other weights you want to use for the galaxies and randoms, 
                                                                                                                      # we set them to be 1 for simplify.
# 2. if nbar is calculated on shell: 
out_dir,wfkpS, NormS  = Pkest_Fun(kmin,kmax,nk,nx,ny,nz,xmin,xmax,ymin,ymax,zmin,zmax,
                                ra ,dec ,zm ,nb_S            ,FKP,
                                raR,decR,zmR,nbR_S*D2R_ratio,
                                np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,
                                OmegaM,OmegaA,Hub,np.nan, 
                                np.nan,PS_type='den',PS_multi=PS_multi,
                                file_dir=file_dir+'PSden_shell', wd=1., wdR=1.)
```

<font color='brown'><font size="4.9">Section 3.1 Ross Method</font></font>

This method only works for $\ell=0$ auto-density and auto-momentum power spectrum. 

This method is from Ross et. al 2013: https://arxiv.org/abs/1208.1491 .

Call the following function:

<font color='green'><font size="2">
WFij, Kmod = ConvMat_Ross( Kmea, kminc, kmaxc, nkbinc, WF_rand, Ncosa, epsilon_par )     
</font></font>  

where:
- Kmea : the k-bin of the measured power spectrum of galaxies. 
- kminc, kmaxc, nkbinc : the k-bin corresponding to the model power spectrum. 
- WF_rand=[kR, PkR] : kR and PkR are the k-bin and power spectrum of randoms.  
- Ncosa : intigration precision of angular $\mu=\cos\theta$.
- epsilon_par : intigration precision of $k$.
- WFij : output of window function concolution matrix. 
- Kmod : output of k-bin corresponding to the model power spectrum.

```python
# seting PS_type to compute window function convolution for density or momentum auto power spectrum, seperately!
PS_type     = 'den'      # 'mom' #             

# 1 read k-bin values of the measured PS of data:
infile=np.loadtxt(file_dir+'PS'+PS_type+'_grid');
k_data=infile[:,1]; NKs= len(k_data)
    
# 2 read PS of randoms: 
infile=np.loadtxt(file_dir+'PSrand_'+PS_type);
k_rand=infile[0:len(infile[:,0])-1,1];Pk_rand=infile[0:len(infile[:,0])-1,2];k_rand[0]=0.0;Pk_rand[0]=1.0
ind=np.where(Pk_rand>0.0);k_rand =k_rand[ind];Pk_rand =Pk_rand[ind]
ind=np.where(Pk_rand>0.0);k_rand =k_rand[ind];Pk_rand =Pk_rand[ind]
    
# 3 extend the ps of randoms:
plt.figure(1,figsize=(4,3))
ktemp,Pktemp=PS_extender(k_rand[1:len(k_rand)],Pk_rand[1:len(k_rand)],100,len(k_rand))
k_rands =np.zeros(1+len(ktemp));Pk_rands=np.zeros(1+len(ktemp))
k_rands[0]=0.0;Pk_rands[0]=1.0
k_rands[1:len(k_rands)]=ktemp;Pk_rands[1:len(k_rands)]=Pktemp
k_rand=k_rands
Pk_rand=Pk_rands
ind=np.where(Pk_rand>0.0)
k_rand =k_rand[ind]
Pk_rand =Pk_rand[ind]
plt.figure(3,figsize=(3,2.5))
fsiz=10;tsiz=8
plt.loglog(k_rand, Pk_rand,marker='',color='r')
plt.xlabel('$k$ [h Mpc$^{-1}$]',fontsize=fsiz);
plt.ylabel('$P_r(k)$',fontsize=fsiz) ;plt.title('Power Spectrum of randoms')
plt.show()     
     
# 4 setting parameters:
Ncosa              = 200              # intigration precision for Ross method only.
epsilon_par        = [0.0,0.8,2000]   # intigration precision and intervel for Ross method only.
Kmea               = k_data[0:NKs-1]  # Ki is for the output, which corresponds to measured PS:
kminc,kmaxc,nkbinc = np.min(Kmea), 0.41, 300   # k_mod bins: the kbin of model PS. 
WF_rand            = [k_rand,Pk_rand] # PS of randoms
    
# 5: The WF matrix WFij is to convert the input model PS(kj) 
# to the output PS(ki):
WFij,Kmod=ConvMat_Ross(Kmea,kminc,kmaxc,nkbinc,WF_rand,Ncosa,epsilon_par)  
np.save(file_dir+'conv'+PS_type+'_'+'Ross',np.array([WFij,Kmea,Kmod,WF_rand],dtype=object),allow_pickle=True)  

# 6: make plots: 
conv,kd,kc,tmp   =np.load(file_dir+'conv'+PS_type+'_'+'Ross'+'.npy',allow_pickle=True)
Nkd=len(kd)//4; indkd=np.array([0.5*Nkd,Nkd,1.5*Nkd,2*Nkd,2.5*Nkd,3*Nkd,3.5*Nkd,4*Nkd],dtype=int)-1        
Nkc=len(kc)//4; indkc=np.array([0.5*Nkc,Nkc,1.5*Nkc,2*Nkc,2.5*Nkc,3*Nkc,3.5*Nkc,4*Nkc],dtype=int)             
plt.figure(3,figsize=(4,3))
fsiz=15;tsiz=8
im   = plt.pcolor( conv   ,    cmap='RdBu_r'   )
cbar = plt.colorbar() ;plt.clim(np.min(conv),  np.max(conv));
cbar.ax.tick_params(labelsize=tsiz); cbar.set_label('$W(k_{mod},k_{mea})$',fontsize=fsiz)  ;
plt.tick_params(axis = 'both', which = 'major', labelsize =tsiz); 
plt.xlabel('$k_{mod}$ [h Mpc$^{-1}$]',fontsize=fsiz);
plt.ylabel('$k_{mea}$ [h Mpc$^{-1}$]',fontsize=fsiz) ;
plt.xticks(indkc,np.round(kc[indkc],3));
plt.yticks(indkd,np.round(kd[indkd],3));
plt.show()     
```

where <font color='green'>VpRand_Fun</font>   generates fake peculiar velocities (and errors) for randoms, <font color='green'>Prep_winFUN</font>   grid the randoms, <font color='green'>ConvMat_CosmPS</font>   calculate convolution matrix.
- logd_data : log-sitance ratio of galaxies.
- cz_data,cz_rand : redshift in km/s for galaxies and randoms, respectively. 
- rand_Ratio : the number ratio between randoms and galaxies = 1/D2R_ratio. 
- NczBin : number of redshift-bins to cz_data. 
- xR,yR,zR,vpR,evpR,wR : the (x,y,z) positions of randoms, velocities and errors and any other weights of the randoms, respectively. 
- kmin,kmax,nkbin : the k-bin of the measured power spectrum of galaxies. 
- kmaxc,nkbinc : the k-bin corresponding to the model power spectrum. 
- nx,ny,nz,lx,ly,lz: grid parameters.  
- weigridgal,wingridgal : the grid of weighted galaxy density field and un-weighted galaxy density field. Outputs of 'Prep_winFUN'.
- weigridpv ,wingridpv : the grid of weighted galaxy velocity field and un-weighted galaxy velocity field.Outputs of 'Prep_winFUN'.
- conv : output of window function concolution matrix. 
- kmea : output of k-bin corresponding to the measured power spectrum.
- kmod : output of k-bin corresponding to the model power spectrum.
- kmod_scale= 'lin', 'log': the bin scale for kbins of model power spectrum, linear, log scales. 


<font color='red'><font size="3.9">Important Tips:</font></font> 
- The nx,ny,nz,nkbinc should not too large, unless the number of random points are large enough to produce a smooth grid for window function grid. Otherwise, <font color='red'>the convolved PS (especially the den-PS) will systmaticly lower than the unconvolved PS.</font>  
- The kmaxc should be slightly larger then kmax, but if kmaxc is much larger then kmax, the convolved PS will has a shift compare to the unconvolved PS. So do not set kmaxc too large, unless the random catalogue is big enough. 
- If 'nkbinc' and/or 'nkbin' is too large, it will cause some numerical precision problems unless the random catalogue is big enough, otherwise, their will be too may '<font color='red'>-1.</font>' values in the matrix 'conv',  '-1.' (euqivlent to 'np.nan' in our code) indicates the k-bin is empty. 

```python
# 1. stetting parameters:
OmegaM      = 0.3151
OmegaA      = 1.0-OmegaM
Hub         = 100.
sigv        = 300.
FKP         = 5000.
FKPv        = 10.**10    
    
nx,ny,nz    = 140  , 140  , 140       # grid parameters
lx,ly,lz    = 700.0, 700.0, 700.0  
    
kmin ,kmax ,nkbin  = 0.0,0.35, 66 
kminc,kmaxc,nkbinc = 0.0,0.41, 170
 
randv_Ratio,randg_Ratio = 9.95367,9.95367  # the number ratio between randoms and galaxies.
                                           # see sectioin 2.2.1.
                                           # randv_Ratio: the number ratio between randoms and PVs
                                           #randg_Ratio: the number ratio between randoms and galaxies
                                           # since we use single catalogue for density and velocity fields
                                           # we have: randv_Ratio=randg_Ratio=1/D2R_ratio
                    
#2 read randoms of den-gal sample:
from PSestFun import *
# load data and randoms
raR,decR,zmR     = np.load(file_dir+"rand_ori.npy", allow_pickle=True);                                                         
ra,dec,zm,vp,evp = np.load(file_dir+"data_ori.npy", allow_pickle=True); 
D2R_ratio=len(ra)/len(raR)
ravR,decvR, zmvR = raR,decR,zmR
weit = 1. # setting the weight of galaxies to be 1 for simplify. 
#calculate nbar on grid: 
delt_z     = 0.002  # the thickness of shell. 
SurveyArea = 4.*np.pi*0.413159  # the area of sky coverage of the survey. 
nbR_S      = nbarS_Fun(delt_z,zmR,weit,SurveyArea,OmegaM,OmegaA,Hub)# nbar of randoms. 
nbRnorm    = nbR_S * len(ra)/len(raR)
# weight density:
randg_weit= 1./(   1.+nbRnorm*FKP)
# generating velocity errors for random points 
logd_data = Vp_to_logd(vp,zm,OmegaM) 
NczBin    = 50 # bins of redshift, for calculating the scatter of PVs in each redshift bin.
vpR,evpR  = VpRand_Fun(logd_data,zm*LightSpeed,zmR*LightSpeed,NczBin,OmegaM)
# weight velocity:
nbvRnorm=nbRnorm
randv_weit= 1./(evpR**2+sigv**2+ nbvRnorm*FKPv) 

#3 convert to cartician, density data:
randg_x,randg_y,randg_z  = Sky2Cat(raR,decR,zmR,OmegaM , OmegaA ,Hub,nbin=20000,redmax=2)   
del raR,decR,zmR,nbRnorm
gc.collect()# save memory
#convert to cartician, velocity data:
randv_x,randv_y,randv_z  = Sky2Cat(ravR,decvR, zmvR ,OmegaM , OmegaA ,Hub,nbin=20000,redmax=2)
del ravR,decvR,zmvR,nbvRnorm
gc.collect()                     

# 4 grid galaxy data and PV data:   
wingridgal,weigridgal = Prep_winFUN(randg_Ratio,FKP,randg_x,randg_y,randg_z,
                                    np.nan,1.,np.nan,nx,ny,nz,lx,ly,lz,'den')  
wingridpv,weigridpv   = Prep_winFUN(randv_Ratio,FKPv,randv_x,randv_y,randv_z,
                                    evpR  ,1.,sigv  ,nx,ny,nz,lx,ly,lz,'mom')

# 5 calculate matrix:  
for i in range(3):
    if(i==0): PS_type     =  'crs'
    if(i==1): PS_type     =  'den' 
    if(i==2): PS_type     =  'mom'  
    print('\n PS_type=',PS_type)          
    if(PS_type=='den'):
        conv,kd,kc   =  ConvMat_CosmPS(kmin,kmax,nkbin,kmaxc,nkbinc,nx,ny,nz,lx,ly,lz,
                                      weigridgal,wingridgal ,weigridgal,wingridgal,PS_type,'log') 
    if(PS_type=='mom'):
        conv,kd,kc   =  ConvMat_CosmPS(kmin,kmax,nkbin,kmaxc,nkbinc,nx,ny,nz,lx,ly,lz,
                                      weigridpv ,wingridpv  ,weigridpv ,wingridpv ,PS_type,'log') 
    if(PS_type=='crs'):
        conv,kd,kc   =  ConvMat_CosmPS(kmin,kmax,nkbin,kmaxc,nkbinc,nx,ny,nz,lx,ly,lz,
                                      weigridgal,wingridgal ,weigridpv ,wingridpv ,PS_type,'log') 
    np.save(file_dir+'conv'+PS_type+'_'+'CosmPS',np.array([conv,kd,kc],dtype=object),allow_pickle=True)
```

In this part, we show some examples for the calculations of the redshift-space galaxy power spectrum models. The code of the models are developed based on the perturbation theory of the following papers:\
V12: https://arxiv.org/pdf/1207.0839 \
V13: https://arxiv.org/pdf/1308.6294 \
O14: https://arxiv.org/pdf/1312.4214v2 \
S14: https://arxiv.org/abs/1405.1447 \
H19: https://arxiv.org/pdf/1906.02875 \
Q19: https://arxiv.org/abs/1906.02874 \
Q24: https://arxiv.org/abs/2411.09571 \
Please <font size="4"><font color='red'>cite</font></font> the above papers if you want to use our code to compute the models of galaxy power spectrum. 


The module we have use to model the power spectrum is <font color='brown'>PSmodFun.py</font>, to load the module, please run ' <font color='green'> from PSmodFun import * </font>'. In this notebook, we will show how to compute:\
 1: the redshift-space galaxy power spectrum model $P(k,\mu)$; \
 2: the power spectrum multipole model $P_{\ell}(k )$; \
 3: the window function convolved power spectrum multipole model $P^c_{\ell}(k )$. \
We have the auto-density power spectrum $P^{\delta}$, the auto-momentum power spectrum $P^{p}$ and the density-momentum cross-power spectrum $P^{\delta p}$ to compute.
 
\
<font color='brown'><font size="4.9">Section 4.2 The Parameters of the Models</font></font>

<font color='brown'><font size="3.9">4.2.1 The Parameters</font></font>

The free parameters of the power spectrum models are listed below. The keywords of the parameters in the code are:

- <font color='blue'>fsigma8</font>    : $f\sigma_8$ the growth rate. It should be larger than zero, $f\sigma_8\in(0,+\infty)$.  
- <font color='blue'>epsilon</font>    : $\epsilon$ a shift on the velocities of galaxies. This parameter is only for the Galilean-transformed auto-momentum and cross-power spectrum. The auto-density power spectrum is not affected by Galilean-transformation. $\epsilon\in(-\infty,+\infty)$, the unite is $km/s$.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<font color='lightpink'> Density field parameters</font> 
- <font color='blue'>b1sigma8</font>   : $b_1\sigma_8$ the first-order biasing parameter of the galaxy density field. It should be larger than zero, $b_1\sigma_8\in(0,+\infty)$.
- <font color='blue'>b2sigma8</font>   : $b_2\sigma_8$ the second-order local biasing parameter of the galaxy density field, and $b_2\sigma_8\in(-\infty,+\infty)$. It can smaller than zero.  
- <font color='blue'>bssigma8</font>   : $b_s\sigma_8$ the second-order nonlocal biasing parameter of the galaxy density field, $b_s\sigma_8\in(-\infty,+\infty)$. It can smaller than zero too. It comes from the tidal tensor. If setting bssigma8='anal' when fitting the models to the measurments, it will be computed using the local Lagrangian relations, rather than treating it as a free parameter. We strongly recommend the users to set bssigma8='anal' when fitting the models. But anyway, it is up to the users.
- <font color='blue'>b3nlsigma8</font> : $b_{3nl}\sigma_8$ the third-order nonlocal biasing parameter of the galaxy density field, $b_{3nl}\sigma_8\in(-\infty,+\infty)$.  If setting b3nlsigma8='anal', it will be computed using the local Lagrangian relations. 
- <font color='blue'>sigmav1_squre</font>   : $\sigma^2_{v1}$ the velocity dispersion parameter of the galaxy density field, correspongding to the loop terms $P_{02}$, $P_{12}$, $P_{22}$ and the vector part of $P_{13}$. It should be larger or euqal than zero, $\sigma^2_{v1}\in[0,+\infty)$. The unit is $km^2/s^2$.
- <font color='blue'>sigmav2_squre</font>   : $\sigma^2_{v2}$ the velocity dispersion parameter of the galaxy density field, correspongding to the loop terms $P_{03}$ and the scalar part of$P_{13}$. $\sigma^2_{v2}\in[0,+\infty)$. The $\ell\ge2$ power spectrum multipole models are sensitive to this parameter. But if the users only want to fit the $\ell=0$ model,  they may want to set its value to be the same as $\sigma^2_{v1}$, i.e. sigmav2_squre=sigmav1_squre. Anyway, it is up to the users. The unit is $km^2/s^2$.
- <font color='blue'>sigmav3_squre</font>   : $\sigma^2_{v3}$ the velocity dispersion parameter of the galaxy density field, correspongding to the loop term $P_{04}$.  $\sigma^2_{v3}\in[0,+\infty)$. The power spectrum models are not sensitive to this parameter, the users may want to set it to be sigmav3_squre=sigmav1_squre too. The unit is $km^2/s^2$. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<font color='lightpink'>Momentum field parameters</font>  
- <font color='blue'>b1vsigma8</font>   : $b_{v1}\sigma_8$ the first-order biasing parameter of the galaxy velocity field, and $b_1\sigma_8\in(0,+\infty)$.
- <font color='blue'>b2vsigma8</font>   : $b_{v2}\sigma_8$ the second-order local biasing parameter of the galaxy velocity field, and $b_2\sigma_8\in(-\infty,+\infty)$.  The model of the momentum power spectrum is not sensitive to this parameter, the users can set its value to be the same as the value of the second-order biasing parameter of the density field $b_{2}\sigma_8$, i.e. b2vsigma8=b2sigma8. 
- <font color='blue'>bsvsigma8</font>   : $b_{vs}\sigma_8$ the second-order nonlocal biasing parameter of the velocity field, $b_{vs}\sigma_8\in(-\infty,+\infty)$. If setting bsvsigma8='anal', it will be computed using the local Lagrangian relations. The model of the momentum power spectrum is not sensitive to this parameter, the users can set it to be the same as the density field too, i.e. bsvsigma8=bssigma8 or set it to be bsvsigma8='anal'. 
- <font color='blue'>b3nlvsigma8</font> : $b_{v3nl}\sigma_8$ the third-order nonlocal biasing parameter of the velocity field, $b_{v3nl}\sigma_8\in(-\infty,+\infty)$. If setting b3nlvsigma8='anal', it will be computed using the local Lagrangian relations. The momentum power spectrum model is not sensitive to this parameter, similarly, the users can set it to be b3nlvsigma8=b3nlsigma8 or b3nlvsigma8='anal'. 
- <font color='blue'>sigmavv1_squre</font>   : $\sigma^2_{vv1}$ the velocity dispersion parameter of the velocity field,   $\sigma^2_{vv1}\in[0,+\infty)$. The unit is $km^2/s^2$.
- <font color='blue'>sigmavv2_squre</font>   : $\sigma^2_{vv2}$ the velocity dispersion parameter of the velocity field, $\sigma^2_{vv2}\in[0,+\infty)$. Similarly, the models of the $\ell\ge2$ power spectrum multipoles are sensitive to this parameter. But if the users only want to fit the $\ell=0$ term, they may set it to be sigmavv2_squre=sigmavv1_squre. The unit is $km^2/s^2$.
- <font color='blue'>sigmavv3_squre</font>   : $\sigma^2_{vv3}$ the velocity dispersion parameter of the velocity field, $\sigma^2_{vv3}\in[0,+\infty)$. This parameter is not sensitive in the fit, the users may set it to be sigmavv3_squre=sigmavv1_squre. The unit is $km^2/s^2$.

To compute the auto-density power spectrum model, you only need the fsigma8 and the 7 density field parameters. \
To compute the auto-momentum power spectrum model, you only need the fsigma8 and the 7 momentum field parameters. \
To compute the cross-power spectrum model, you need 11 parameters:\
[$f\sigma_8$,$b_1\sigma_8$,$b_2\sigma_8$,$b_s\sigma_8$,$b_{v1}\sigma_8$,$b_{v2}\sigma_8$,$b_{vs}\sigma_8$,$b^c_{3nl}\sigma_8$,$\sigma^2_{cv1}$,$\sigma^2_{cv2}$,$\sigma^2_{cv3}$]\
where  $b^c_{3nl}\sigma_8$ can either from momentum field or density field, [$\sigma^2_{cv1}$,$\sigma^2_{cv2}$,$\sigma^2_{cv3}$] can either from density field or momentum field.\
If  $b^c_{3nl}\sigma_8$ is taken from density field, [$\sigma^2_{cv1}$,$\sigma^2_{cv2}$,$\sigma^2_{cv3}$] should be taken from  momentum field, i.e. [$b_{3nl}\sigma_8$, $\sigma^2_{vv1}$, $\sigma^2_{vv2}$, $\sigma^2_{vv3}$].\
While if  $b^c_{3nl}\sigma_8$ is taken from momentum field, [$\sigma^2_{cv1}$,$\sigma^2_{cv2}$,$\sigma^2_{cv3}$] should  be taken from density field, i.e. [$b_{v3nl}\sigma_8$, $\sigma^2_{v1}$, $\sigma^2_{v2}$, $\sigma^2_{v3}$]. The users can comparing the model cross power to the measured cross power of mocks, to see which one can give better fit results, i.e more close to fiducial parameters. \
To compute the Galilean-transformed auto-momentum and cross-power spectrum, you need one more free parameter $\epsilon$.


<font color='brown'><font size="3.9">4.2.2 Some Suggestions About the Choice of the Free Parameters</font></font>

The power spectrum models are not sensitive to all these parameters in a equal level. So, in a practice level, we can reduce some parameters.  

Fitting $P^{\delta}_{0}+P^{p}_{0}$ : \
We can set ($f\sigma_8$,  $b_1\sigma_8$, $b_2\sigma_8$, $\sigma^2_v$) as free parameters. 
In the meanwhile, setting bssigma8='anal', bsvsigma8='anal', b3nlsigma8='anal', b3nlvsigma8='anal'. All the other  biasing parameters of the momentum field are set to be the same as the density field.  Setting all the  velocity dispersion parameters to be the same vale $\sigma^2_v$.  Therefore, we will only have four free parameters to make the models  flexible enought to fit the measurements.
 
Fitting $P^{\delta}_{0}+P^{p}_{0}+P^{\delta p}_{1}$: \
We can set ($f\sigma_8$,  $b_1\sigma_8$, $b_2\sigma_8$, $b_{3nl}\sigma_8$, $\sigma^2_v$) as free parameters. 
In the meanwhile, setting bssigma8='anal', bsvsigma8='anal'. All the other biasing parameters of the momentum field are set to be the same as the density field.  Setting all the  velocity dispersion parameters to be the same vale $\sigma^2_v$.  Therefore, we will only need five free parameters to make the models flexible enought to fit the measurements.

Fitting $P^{\delta}_{02}+P^{p}_{0}+P^{\delta p}_{1}$ : \
We can set ($f\sigma_8$,  $b_1\sigma_8$, $b_2\sigma_8$, $b_{3nl}\sigma_8$, $b_{v1}\sigma_8$, $\sigma^2_{v1}$, $\sigma^2_{v2}$, $\sigma^2_{vv}$) as free parameters. 
In the meanwhile, setting bssigma8='anal', bsvsigma8='anal', b3nlvsigma8='anal'. All the other biasing parameters of the momentum field are set to be the same as the density field. Setting $\sigma^2_{v3}$ of the density field to be sigmav3_squre=sigmav1_squre.  Setting all the  velocity dispersion parameters of the momentum field to be the same vale $\sigma^2_{vv}$.  Therefore, we will have 8 free parameters to make the models flexible enought to fit the measurements. 
 
Fitting $P^{\delta}_{02}+P^{p}_{02}+P^{\delta p}_{1}$ : \
We can set ($f\sigma_8$,  $b_1\sigma_8$, $b_2\sigma_8$, $b_{3nl}\sigma_8$, $b_{v1}\sigma_8$, $b_{v2}\sigma_8$, $\sigma^2_{v1}$, $\sigma^2_{v2}$, $\sigma^2_{vv1}$, $\sigma^2_{vv2}$) as free parameters. 
In the meanwhile, setting b3sigma8='anal', b3vsigma8='anal', sigmav3_squre=sigmav1_squre, b3nlvsigma8='anal', sigmavv3_squre=sigmavv1_squre.  Therefore, we will have 10 free parameters to make the models flexible enought to fit the measurements. 



<font color='red'><font size="3.9">Important Tips:</font></font> \
We strongly recommend the users to set bssigma8='anal' and bsvsigma8='anal' when fittinng the models, rather than setting them as free parameters. \
The choice of the free parameters are completely up to the users. The users should compare the model to their measurements of mocks to figure out that how many parameters are required to make the model flexible enough to reproduce the fiducial  $f\sigma_8$ of the mocks.


<font color='brown'><font size="4.9">Section 4.3 Calculating the Redshift-Space Power Spectrum Model $P(k,\mu)$</font></font>

<font color='brown'><font size="3.9">4.3.1 The Redshift-Space Power Spectrum Model $P(k,\mu)$</font></font>

To calculate the redshift-space power spectrum model $P(k,\mu)$, the users can simply copy/paste and run the following 4 lines: 
 
<font color='green'><font size="2">
from PSmodFun import *  \
PT, Sig8_fid, outdir&nbsp;= PSloop_Fun( KmodLim,Hub,ombh2,omch2,ns,As,file_dir) \
az&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;= 1./(1.+RsFeff)\
Hz&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;= Hub * Ez(RsFeff, OmegaM, OmegaA, 0., -1., 0., 0.)\
Dz&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;= Dz_Fun(RsFeff, OmegaM, OmegaA, Hub)         
k, mu, Pkmu &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;= Pkmod_Fun(parm, Sig8_fid, PT, PS_type, az, Hz,Dz,False,GaliTrans)  
</font></font>
 

where the keyword: \
PS_type = 'den', 'mom' or 'crs' or 'crsv' for auto-density, auto-momentum or cross- power spectrum, respectively. If using 'crs', in the cross power model, the $b_{3nl}\sigma_8$ will be taken from momentum field, while velocity dispersion parameters will be taken from density field, i.e. using [$b_{v3nl}\sigma_8$, $\sigma^2_{v1}$, $\sigma^2_{v2}$, $\sigma^2_{v3}$] for cross-power model. If using 'crsv', in the cross power model, the $b_{3nl}\sigma_8$ will be taken from momentum field, while velocity dispersion parameters will be taken from density field, i.e. using [$b_{3nl}\sigma_8$, $\sigma^2_{vv1}$, $\sigma^2_{vv2}$, $\sigma^2_{vv3}$] for cross-power model. The users need to use mocks to find out which one can give better fit results.       
az: scale-factor $a(z)$.\
Dz: growth factor $D(z)$.\
Hz: Hubble parameter $H(z)$.\
Hub: Hubble constant at z=0.\
RsFeff: effective redshift of your data.\
Ez: This is an internal function of our code, you can directly call it. \
GaliTrans: True or False, if GaliTrans=True, return the Galilean-transformed power spectrum. The default value if False.

The parameters of the model are encoded in the following order:\
parm = [ fsigma8 ,  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;     b1sigma8 &nbsp;&nbsp;, b2sigma8 &nbsp;&nbsp;, bssigma8&nbsp;&nbsp;, b3nlsigma8 ,  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;      b1vsigma8 , b2vsigma8 , bsvsigma8 , b3nlvsigma8 ,   
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;      sigmav1_squre &nbsp;&nbsp;, sigmav2_squre&nbsp;&nbsp;,  sigmav3_squre ,    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;      sigmavv1_squre , sigmavv2_squre , sigmavv3_squre, (epsilon) ]  . \
Therefore, <font color='red'><font size="3">the users need to input the parameters in a correct order as shown in the above</font></font>. If 'GaliTrans=True', the users need to set a value for 'epsilon'.      


In the following, we will show the computation of the model power spectrum step by step. We do not consider the Galilean-transformation. 

```python
import numpy as np
import matplotlib.pyplot as plt
from PSmodFun import * # load the module of the model power spectrum  

# This is the directory to store your input and output data. This string should end with '/'.
file_dir = "../Data/"   
```

<font color='blue'>Step 1Ôºö Calculating the Loop terms: </font>

The 'camb' is used in the following <font color='green'>PSloop_Fun</font> to calculate the real-space linear matter density power spectrum $P_L(k)$. The 'camb' can be downloaded from https://pypi.org/project/camb/ . 

```python
# The cosmological parameters used to calculate the linear matter power spectrum PL(k) of CAMB.
# They should be the same as the fiducial values of the mocks or simulations you want to use.
# For the abacus_cosm000 simulation and mocks, they are in: 
# https://abacussummit.readthedocs.io/en/latest/cosmologies.html
# or Table.1 and 5 of https://www.aanda.org/articles/aa/pdf/2020/09/aa33910-18.pdf
# The fiducial sigma8_fid_sim =0.811355
OmegaM    = 0.315192
OmegaA    = 1.- OmegaM
hb        = 0.67360
Hub       = 100.
ombh2     = 0.02237
omch2     = 0.12000
ns        = 0.96490
As        = 2.1011980518085315e-09
# The 'As' parameter is important, users should adjust it to re-produce the fiducial 'sigma8_fid_sim=0.811355' 
# of the mocks by using the following relation:
# 'As = As*sigma8_fid_sim/sigma8_fid_integ'
# where 'sigma8_fid_integ' is the output value on the screen. 
# Adjust 'As' untill 'sigma8_fid_integ' euqals to 'sigma8_fid_sim'.

#The maximume value of k used to compute PL(k).
KmodLim = 1.  # h/Mpc. Yes, this is a very large value for power spectrum.
NKmodbin= 437 # the number of k-bin for PL(k) of camb.

# You only need to calculate the loop terms 'PT' once for a given set of cosmological parameters, 
# and the results will be automatically saved into "  outdir = file_dir+'INTEG_PL.npy' " after the calculation been finished.
# It will take around 1h30mins to make these calculations on NERSC (but only 20mins on MacBook Pro), please be patience.  
PT, Sig8_fid, outdir = PSloop_Fun( KmodLim,NKmodbin,Hub*hb,ombh2,omch2,ns,As,file_dir) 
print('\n The loop terms are saved in:', outdir, '\n')
```

In the output loop term 'PT', the first colume is the k values of the linear matter power spectrum:\
klin=PT[0,:] \
the second colume is the corresponding linear matter power spectrum:\
Plin=PT[1,:] \
In the above 'PSloop_Fun', we use CAMB to calculate klin and Plin.

```python
klin=PT[0,:]
Plin=PT[1,:]
plt.figure(1,figsize=(4,3))
fsiz=15;tsiz=10
plt.loglog(klin,Plin,c='r');
plt.xlabel('k [h/Mpc]',fontsize=fsiz);plt.ylabel('$P_{lin}(k)$',fontsize=fsiz);plt.tick_params( labelsize =tsiz);
plt.title('Linear matter power spectrum CAMB')
plt.show()
```

<font color='blue'>Step 2Ôºö Calculating scale factor $a(z)$ and growth factor $D(z)$ and Hubble parameter $H(z)$ at effective red-shift $z=z_{eff}$: </font>

```python
RsFeff = 0.05   # An example effective redshift of a survey. It is not for DESI, it should be updated later for DESI PV. 
az     = 1./(1.+RsFeff)
Hz     = Hub * Ez(RsFeff, OmegaM, OmegaA, 0., -1., 0., 0.) # just use Hub = 100. 
Dz     = Dz_Fun(RsFeff, OmegaM, OmegaA )    
```

<font color='blue'>Step 3Ôºö Setting the parameters of power spectrum model: </font>

```python
# Here we just manually set these parameters, they are not fit results from any real surveys.
# growth rate:------------------
fsigma8        = 0.40946855 
# density field:---------------- 
b1sigma8       = 1.12568620 
b2sigma8       = 1.17890514 
bssigma8       = 'anal' 
b3nlsigma8     = 0.4335061 
sigmav1_squre  = 40.2873740 
sigmav2_squre  = 156.669623 
sigmav3_squre  = 1.8238754 
# momentum field:--------------- 
b1vsigma8      = 0.80151994 
b2vsigma8      = b2sigma8
bsvsigma8      = 'anal'
b3nlvsigma8    = 'anal'
sigmavv1_squre = 3.1344342 
sigmavv2_squre = 12.1289577
sigmavv3_squre = 1.27892345
#-------------------------------
# The list of parameters of power spectrum: 
# As mentioned above, the parameters should be listed in the correct order:
parm = [fsigma8  ,  
        b1sigma8 , b2sigma8 , bssigma8 , b3nlsigma8   ,  
        b1vsigma8, b2vsigma8, bsvsigma8, b3nlvsigma8  ,   
        sigmav1_squre , sigmav2_squre  , sigmav3_squre,    
        sigmavv1_squre, sigmavv2_squre , sigmavv3_squre] 
#-------------------------------
# The list of parameters of Galilean-transformed power spectrum: 
# As mentioned above, the parameters should be listed in the correct order:
epsilon = 200. #km/s
parmG = [fsigma8  ,  
        b1sigma8 , b2sigma8 , bssigma8 , b3nlsigma8   ,  
        b1vsigma8, b2vsigma8, bsvsigma8, b3nlvsigma8  ,   
        sigmav1_squre , sigmav2_squre  , sigmav3_squre,    
        sigmavv1_squre, sigmavv2_squre , sigmavv3_squre, epsilon]   
```

<font color='blue'>Step 4Ôºö Calculating power spectrum model $P(k,\mu)$: </font>

```python
krsd,mu,Pkmodrsd_den = Pkmod_Fun(parm, Sig8_fid, PT, 'den', az,Hz,Dz,False) # density  auto-power spectrum. 
krsd,mu,Pkmodrsd_mom = Pkmod_Fun(parm, Sig8_fid, PT, 'mom', az,Hz,Dz,False) # momentum auto-power spectrum.
krsd,mu,Pkmodrsd_crs = Pkmod_Fun(parm, Sig8_fid, PT, 'crs', az,Hz,Dz,False) # density-momentum cross-power spectrum.the b3nl will be taken from momentum field, while velocity dispersion parameters will be taken from density field 
krsd,mu,Pkmodrsd_crsv= Pkmod_Fun(parm, Sig8_fid, PT, 'crsv', az,Hz,Dz,False) #density-momentum cross-power spectrum.the b3nl will be taken from density  field, while velocity dispersion parameters will be taken from momentum field

# Galilean-transformed power spectrum: 
krsd,mu,Pkmodrsd_momG = Pkmod_Fun(parmG, Sig8_fid, PT, 'mom', az,Hz,Dz,False, True) # Galilean-transformed momentum auto-power spectrum.
krsd,mu,Pkmodrsd_crsG = Pkmod_Fun(parmG, Sig8_fid, PT, 'crs', az,Hz,Dz,False, True) # Galilean-transformed cross-power spectrum. Just taking 'crs' as example. You can try 'crsv' by yourself.
```

\
<font color='brown'><font size="3.9">4.3.2 Convert the 2D Power Spectrum $P(k,\mu)$ to 1D Power Spectrum $P(k)$</font></font>

We can convert the 2D power spectrum $P(k,\mu)$ to 1D power spectrum $P(k)$ by integral $P(k,\mu)$ over $\mu$:
$$P(k)=\int^1_0P(k,\mu)d\mu$$
We can use the Simpson's rule:

<font color='brown'><font size="4.9">Section 4.6 Comparing the Perturbation Models to Kaiser Formulars</font></font>

We compare our perterbation (non-linear) model power spectrum to the linear model power spectrum - Kaiser formulars:\
density-PS: $$P(k)=(b_1+f\mu^2)^2D^2_gP_L(k)$$
momentum-PS: $$P(k)=(\frac{aHf\mu}{k})^2D^2_pP_L(k)$$
cross-PS: $$P(k)=(b_1+f\mu^2)(\frac{aHf\mu}{k})D_gD_pP_L(k)$$
where $P_L(k)$ is the linear matter power spectrum from CAMB, and the damping terms are:\
density field: $$D_g=[1+0.5(k\mu \sigma_{v,g})^2]^{-1/2}$$
velocity field: $$D_v=sinc(k\sigma_{v,p})=\frac{sin(k\sigma_{v,p})}{k\sigma_{v,p}}$$

<font color='red'><font size="4">Important Tips:</font></font> 
- Setting the higher order parameters (e.g. b2, bs, b3, sigv2 and sigv3 ...) of the non-linear model power spectrum all to be zeros can <font color='red'>NOT</font> make our non-linear linear. In the meanwhile, we should also set the non-linear loop terms to be zeros using <font color='green'>PT[2:,:]=np.zeros(len(PT[0,:]))</font> to linearized our non-linear models. 
- Regarding the <font color='red'>density</font>  power spectrum, the linearized perturbation model only consistent with the Kaiser Formular at redshift zero, i.e. $a(z=0)=1$ or Growthz=1, since the non-linear model density power spectrum is a function of $a(z)$, but in the Kaiser Formular, the density power spectrum is independent of  $a(z)$. So, we only comparing the linearized density power spectrum model to Kaiser formular at z=0 by setting Growthz=1.

For simplify, we set $\sigma_v=0$ in Kaiser formular, therefore, there is not damping.

The function is:\
<font color='green'><font size="2">
from PSmodFun import *  \
r, CFs = CFmodMulti_Fun(r, k, Pmod , l,   kcut, PS_type) 
</font></font> 

where the keywords are: 

PS_type='den', 'mom' or 'crs' for auto-density, auto-momentum or cross- power spectrum, respectively. 

k: k-bins of the model power spectrum.

Pmod :  power spectrum models. 

l :  the  power spectrum multipole order.  

kcut: maximum k value to use in power law extrapolation.

\
Firstly, we need to calculate the power spectrum models:

```python
import numpy as np
import matplotlib.pyplot as plt
from PSmodFun import * # load the module of the model power spectrum  

# This is the directory to store your input and output data. This string should end with '/'.
file_dir =  "../Data/"   

# As mentioned before, we only need to calculate the loop term once, 
# therefore, in here, we directly load the loop terms from file.
sst          = np.load(file_dir+'INTEG_PL.npy',allow_pickle=True)
PT, Sig8_fid = sst  

# Here we just manually set these parameters, they are not fit results from any real surveys.
# growth rate:------------------
fsigma8        = 0.40946855 
# density field:---------------- 
b1sigma8       = 1.12568620 
b2sigma8       = 1.17890514 
bssigma8       = 'anal' 
b3nlsigma8     = 0.4335061 
sigmav1_squre  = 40.2873740 
sigmav2_squre  = 156.669623 
sigmav3_squre  = 1.8238754 
# momentum field:--------------- 
b1vsigma8      = 0.80151994 
b2vsigma8      = b2sigma8
bsvsigma8      = 'anal'
b3nlvsigma8    = 'anal'
sigmavv1_squre = 3.1344342 
sigmavv2_squre = 12.1289577
sigmavv3_squre = 1.27892345
#-------------------------------
# The list of parameters of power spectrum: 
parm = [fsigma8  ,  
        b1sigma8 , b2sigma8 , bssigma8 , b3nlsigma8   ,  
        b1vsigma8, b2vsigma8, bsvsigma8, b3nlvsigma8  ,   
        sigmav1_squre , sigmav2_squre  , sigmav3_squre,    
        sigmavv1_squre, sigmavv2_squre , sigmavv3_squre]

# set k-bins:
Nk           = 3000
kbin         = np.linspace(0.0001,0.42,Nk )

# the pre-factors of model-PS:
OmegaM  = 0.315192
OmegaA  = 1.- OmegaM
Hub     = 100.  
RsFeff = 0.20    
az     = 1./(1.+RsFeff)
Hz     = Hub * Ez(RsFeff, OmegaM, OmegaA, 0., -1., 0., 0.) # just use Hub = 100. 
Dz     = Dz_Fun(RsFeff, OmegaM, OmegaA)    

# calculate Power Spectrum Multipole Models:
# density ps:
Pkmulti_den,kmulti_den, tmp,tmp = PkmodMulti_Fun(parm,False, Sig8_fid, kbin, 
                                                 np.nan,np.nan,np.nan, PT, 'den-024', az,Hz,Dz, np.nan)
# mom ps:
Pkmulti_mom,kmulti_mom, tmp,tmp = PkmodMulti_Fun(parm,False, Sig8_fid, kbin, 
                                                 np.nan,np.nan,np.nan, PT, 'mom-024', az,Hz,Dz, np.nan)
# crs ps:
Pkmulti_crs,kmulti_crs, tmp,tmp = PkmodMulti_Fun(parm,False, Sig8_fid, kbin, 
                                                 np.nan,np.nan,np.nan, PT, 'crs-13' , az,Hz,Dz, np.nan) 
```



\
<font color='brown'><font size="4.9">5.1 Galaxy Correlation Function</font></font>

```python
# k-bins of density power spectrum:
k     = kmulti_den[0*Nk:1*Nk]

# seperation-bins of correlation functions:
r     = np.logspace(np.log10(np.min(15)), np.log10(np.max(250)), 1000)

# density power spectrum multipoles:
Pden0 = Pkmulti_den[0*Nk:1*Nk]
Pden2 = Pkmulti_den[1*Nk:2*Nk]
Pden4 = Pkmulti_den[2*Nk:3*Nk]

# galaxy two-point correlation function multipoles:
r0,Xi0= CFmodMulti_Fun(r, k , Pden0, l=0, kcut=0.1,  PS_type='den') 
r2,Xi2= CFmodMulti_Fun(r, k , Pden2, l=2, kcut=0.1,  PS_type='den') 
r4,Xi4= CFmodMulti_Fun(r, k , Pden4, l=4, kcut=0.1,  PS_type='den') 
 
# make some plots:
plt.figure(1,figsize=(5,4))
plt.plot(r0, r0**2*Xi0, c='r' ,label='$\ell=0$');
plt.plot(r2, r2**2*Xi2, c='g' ,label='$\ell=2$');
#plt.plot(r4, r4**4*Xi4, c='b' ,label='$\ell=4$');
plt.legend(fontsize=18); plt.xticks(fontsize=18); plt.yticks(fontsize=18); 
plt.xlim(10,255);plt.ylim(-60,90);
plt.ylabel(r'$\xi^{\ell}_{gg}(r)$',fontsize=23);plt.xlabel('r [Mpc $h^{-1}$]',fontsize=23)
```

<font color='brown'><font size="4.9">5.2 Velocity Correlation Function</font></font>

```python
# momentum power spectrum l=0:
Pmom0 = Pkmulti_mom[0*Nk:1*Nk]

# velocity two-point correlation function multipoles:
r,psiperp, psipar= CFmodMulti_Fun(r, k , Pmom0, l=0, kcut=0.1, PS_type='mom') 

# make some plots:
plt.figure(2,figsize=(5,4))
plt.plot(r, psiperp, c='r' ,label='$\psi_{||}$');
plt.plot(r, psipar , c='g' ,label='$\psi_{\perp}$');
plt.legend(fontsize=18); plt.xticks(fontsize=18); plt.yticks(fontsize=18); 
plt.xlim(10,255); plt.ylim(-4500,59000);
plt.ylabel('$\psi(r)$',fontsize=23);plt.xlabel('r [Mpc $h^{-1}$]',fontsize=23)
```

```python
# momentum power spectrum l=0:
Pcrs1 = Pkmulti_crs[0*Nk:1*Nk]

# velocity two-point correlation function multipoles:
r,gv_dip= CFmodMulti_Fun(r, k , Pcrs1, l=1, kcut=0.1, PS_type='crs') 

# make some plots:
plt.figure(3,figsize=(4,3))
plt.plot(r, gv_dip , c='g' ); 
plt.xticks(fontsize=18); plt.yticks(fontsize=18); 
plt.xlim(10,255); plt.ylim( 0,100);
plt.ylabel(r'$\xi_{gv}(r)$',fontsize=23);plt.xlabel('r [Mpc $h^{-1}$]',fontsize=23)
```
